```python
import torch
import numpy as np
from plyfile import PlyData, PlyElement
```

```python
splats = torch.load("/home/roman/ba/envgs/data/trained_model/envgs_sedan_50/latest.pt", map_location="cpu")["model"]
```

```python
splats.keys()
```

```python
xyz = splats["sampler.env._xyz"].detach().cpu().numpy().astype(np.float32)
xyz = np.ascontiguousarray(xyz)

centers = xyz.view([("x","f4"), ("y","f4"), ("z","f4")]).reshape(-1)
```

```python
el = PlyElement.describe(centers, "vertex")
```

```python
el
```

```python
PlyData([el]).write('env_xyz.ply')
```
