```python
masks = list(Path("/home/roman/ba/sedan_car_masks").iterdir())
```

```python
normal_dirs = list(Path("/home/roman/ba/datasets/ref_real/sedan_no_bg/normals").iterdir())
```

```python
for normal in tqdm(normal_dirs):
    normal_path = normal / "000000.jpg"
    normal_bgr = cv2.imread(normal_path, cv2.IMREAD_COLOR)

    mask_path = (Path("/home/roman/ba/sedan_car_masks") / normal.name).with_suffix(".png")
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    normal_bgra = cv2.cvtColor(normal_bgr, cv2.COLOR_BGR2BGRA)
    normal_bgra[:, :, 3] = mask
    cv2.imwrite(normal_path.with_suffix(".png"), normal_bgra)
```
